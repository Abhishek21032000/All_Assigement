```python
def video_details(youtube, video_ids):
    request = youtube.videos().list(
    part="snippet,contentDetails,statistics",
    id=','.join(video_ids)
    )
    response = request.execute()

    list_data = response['items']

    data= {
    'thumbnail_url' : [x['snippet']['thumbnails']['default']['url'] for x in list_data],
    'title' : [x['snippet']['title'] for x in list_data],
    'publishedAt' : [x['snippet']['publishedAt'] for x in list_data],
    'view_count' : [x['statistics']['viewCount'] for x in list_data]
    }

    return data


all_data=video_details(youtube, video_ids)
all_data
```
